### D. Cleaning Data

#### Membersihkan Data day_df

##### Memperbaiki kesalahan tipe data untuk kolom dteday dan mengganti tipe data nya menjadi datetime

```python
datetime_columns = ["dteday"]
for column in datetime_columns:
  day_df[column] = pd.to_datetime(day_df[column])
```

```python
day_df.info()
```

okay, cleaning pada dataset day_df sudah siap

#### Membersihkan Data hour_df
